```python
import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as patches
from matplotlib.lines import Line2D
```

```python
df_op = preprocessing.preprocessing_operating_data()
```

```python
df_op['Date and time'] = pd.to_datetime(df_op['Date and time'])

batch_durations = df_op.groupby('Batch')['Date and time'].agg(['min', 'max'])
batch_durations = batch_durations.rename(columns={'min': 'start_time', 'max': 'end_time'})
batch_durations['duration'] = batch_durations['end_time'] - batch_durations['start_time']

df_op = df_op.sort_values(['Batch', 'Date and time'])
df_op['time_diff'] = df_op.groupby('Batch')['Date and time'].diff()

df_op['is_15m'] = (df_op['time_diff'] == pd.Timedelta(minutes=15)).astype(int)
df_op['is_5m'] = (df_op['time_diff'] == pd.Timedelta(minutes=5)).astype(int)

interval_counts = df_op.groupby('Batch')[['is_15m', 'is_5m']].sum()
interval_counts = interval_counts.rename(columns={
    'is_15m': '15_min_intervals', 
    'is_5m': '5_min_intervals'
})

final_summary_table = pd.merge(batch_durations, interval_counts, left_index=True, right_index=True)
final_summary_table['rows'] = df_op.groupby('Batch').size()

freq_mins = (final_summary_table['5_min_intervals'] > final_summary_table['15_min_intervals']).map({True: 5, False: 15})
freq_td = pd.to_timedelta(freq_mins, unit='m')

expected_rows = round(final_summary_table['duration'] / freq_td) + 1

final_summary_table['missing_timestamps'] = (expected_rows - final_summary_table['rows']).astype(int)

display(final_summary_table)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize=(16, 10))
fig.patch.set_facecolor('#f8f9fa') 
ax.set_facecolor('#f8f9fa')

color_15m = '#90A4AE'
color_5m = '#1E88E5'
color_warn_bg = '#FFEBEE'
color_warn_text = '#D32F2F'

y_positions = range(len(final_summary_table))

for y_pos, (batch_num, row) in enumerate(final_summary_table.iterrows()):
    is_high_res = row['5_min_intervals'] > 0
    bar_color = color_5m if is_high_res else color_15m
    
    start = mdates.date2num(row['start_time'])
    end = mdates.date2num(row['end_time'])
    duration = end - start
    
    ax.barh(y_pos, duration, left=start, color=bar_color, height=0.6, zorder=3, edgecolor='white', linewidth=1)
    
    if row['missing_timestamps'] > 10:
        ax.text(end + 5, y_pos, f"⚠️ {row['missing_timestamps']} Missing", 
                va='center', color=color_warn_text, fontweight='bold', fontsize=12,
                bbox=dict(facecolor=color_warn_bg, edgecolor='none', boxstyle='round,pad=0.3'),
                zorder=4)

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))

ax.grid(True, axis='x', color='#e0e0e0', linestyle='--', linewidth=1, zorder=0)
ax.grid(False, axis='y') 
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

ax.set_title('Operating Data: Manufacturing Batch Timeline\n', fontsize=24, fontweight='bold', color='#2c3e50', loc='left')

ax.text(0, 1.02, 'Chronological overview of batch durations, sensor logging resolutions, and data capture anomalies', 
        transform=ax.transAxes, fontsize=15, color='#7f8c8d')

ax.set_xlabel('Production Timeline (Year & Month)', fontsize=16, color='#7f8c8d', labelpad=15, fontweight='bold')
ax.set_ylabel('Manufacturing Batch Number', fontsize=16, color='#7f8c8d', labelpad=15, fontweight='bold')

plt.yticks(y_positions, final_summary_table.index, fontsize=14, color='#34495e', fontweight='bold')
plt.xticks(fontsize=14, color='#34495e')
ax.invert_yaxis()

legend_elements = [
    Line2D([0], [0], color=color_15m, lw=6, label='15-min Resolution'),
    Line2D([0], [0], color=color_5m, lw=6, label='5-min Resolution')
]
ax.legend(handles=legend_elements, loc='upper right', 
          frameon=True, facecolor='white', framealpha=0.9, edgecolor='#CFD8DC', 
          fontsize=13, title='Sensor Logging Frequency', title_fontsize=14)

plt.tight_layout()
plt.savefig("batch_timeline_labelled.jpeg", dpi=300, bbox_inches='tight', facecolor='#f8f9fa')
plt.show()
```

```python
df_p = preprocessing.preprocessing_product_data()
```

```python
batch_product_summary = df_p.groupby('Batch')['Product (g/litre)'].agg(['count', 'mean'])

batch_product_summary = batch_product_summary.rename(columns={
    'count': 'sample_count', 
    'mean': 'mean_product_(g/litre)'
})

display(batch_product_summary)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax1 = plt.subplots(figsize=(16, 8))
fig.patch.set_facecolor('#f8f9fa')
ax1.set_facecolor('#f8f9fa')

batches = batch_product_summary.index.astype(str)
counts = batch_product_summary['sample_count']
means = batch_product_summary['mean_product_(g/litre)']

color_bar = '#CFD8DC'
color_line = '#1E88E5'

ax1.bar(batches, counts, color=color_bar, edgecolor='white', width=0.6, alpha=0.8, label='Sample Count (Volume)')

ax1.set_ylabel('Sample Count (Volume)', fontsize=16, color='#7f8c8d', fontweight='bold', labelpad=15)

ax1.tick_params(axis='y', labelcolor='#546E7A', labelsize=12)
ax1.set_ylim(0, max(counts) * 1.25)

for i, count in enumerate(counts):
    ax1.text(i, count + 3, str(count), ha='center', va='bottom', 
             fontsize=11, color='#7f8c8d', fontweight='bold')

ax2 = ax1.twinx()  
ax2.plot(batches, means, color=color_line, marker='o', markersize=10, linewidth=4, 
         markerfacecolor='white', markeredgewidth=3, label='Mean Product Yield (g/litre)')

ax2.set_ylabel('Mean Product (g/litre)', fontsize=16, color='#7f8c8d', fontweight='bold', labelpad=15)

ax2.tick_params(axis='y', labelcolor=color_line, labelsize=13)
ax2.set_ylim(min(means) * 0.9, max(means) * 1.15)

for i, mean_val in enumerate(means):
    ax2.text(i, mean_val + 0.35, f'{mean_val:.1f}', ha='center', va='bottom', 
             fontsize=12, color=color_line, fontweight='bold')

ax1.set_title('Product Data: Sampling Volume vs. Mean Product (g/litre)\n', 
              fontsize=22, fontweight='bold', color='#2c3e50', loc='left')

ax1.text(0, 1.02, 'Comparison of lab sample counts against mean product yield for all batches in product data', 
        transform=ax1.transAxes, fontsize=15, color='#7f8c8d')

ax1.set_xlabel('Manufacturing Batch Number', fontsize=16, color='#7f8c8d', labelpad=15, fontweight='bold')

ax1.tick_params(axis='x', labelsize=13, rotation=45)

ax1.grid(False) 
ax2.grid(True, axis='y', linestyle='--', alpha=0.6)
for spine in ['top', 'left', 'right', 'bottom']:
    ax1.spines[spine].set_visible(False)
    ax2.spines[spine].set_visible(False)

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right', 
           bbox_to_anchor=(1, 1.15), ncol=2, frameon=True, facecolor='white', 
           framealpha=0.9, edgecolor='#CFD8DC', fontsize=13)

plt.tight_layout()
plt.savefig("batch_yield_combo_improved.jpeg", dpi=300, bbox_inches='tight', facecolor='#f8f9fa')
plt.show()
```
