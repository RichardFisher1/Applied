```python
def moving_avg_forecast(original, window_size):

    forecast = []

    for time in range(len(original) - window_size):
        forecast.append(original[time:time + window_size].mean())

    forecast = np.array(forecast)

    return forecast
```

```python
def exp_moving_avg_forecast(original, alpha = 0.2):

    forecast = []

    forecast.append(original[0])

    for time in range(1, len(original)):
        v = alpha*original[time] + (1 - alpha)*forecast[time - 1]
        forecast.append(v)

    forecast = np.array(forecast)

    return forecast
```

```python
def mse(original, predicted):
    
    n = len(predicted)
    error = 0

    for i in range(n):
        diff = (original[i] - predicted[i])**2
        error = error + diff

    return error / n
```

```python
forecast = moving_avg_forecast(df_plot['Product (g/litre)'].to_numpy(), 10)
```
